# Creación del Dataframe de Pandas y análisis

```python
# Librería de Pandas
import pandas as pd

# Creación del Dataframe
data_path = '../DogRecSys/data/akc-data-latest.csv'
df = pd.read_csv(data_path, encoding='utf-8')

# Encabezado del dataframe
df_head = df.head()

# Información descriptiva (columnas, filas y tipos de datos)
df_info = df.info()

# Resumen descriptivo 
df_desc = df.describe(include='all')

# Conteo de valores nulos
df_nulos = df.isnull().sum()
```
